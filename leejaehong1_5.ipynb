```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
#한글 폰트 설정
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

class StudentScoreAnalysis:
    def __init__(self):
        self.score_columns = ['math_scores', 'english_scores', 'science_scores']

        # 학생 이름 생성
        students = [f'학생{i}' for i in range(1, 21)]

        # 50~100 사이의 난수로 점수 생성
        data = {
        'students': [f'학생{i}' for i in range(1, 21)],
        'math_scores' : np.random.randint(50, 101, 20),
        'english_scores' : np.random.randint(50, 101, 20),
        'science_scores' : np.random.randint(50, 101, 20)
        }
        self.df = pd.DataFrame(data)

    #과목별 평균 점수 막대 그래프 
    def score_mean(self):
        averages = self.df[self.score_columns].mean()
        # 한글 과목명
        subjects = ['수학', '영어', '과학']    

        plt.bar(subjects, averages.values, color='green')
        plt.title('Method 1: 학생별 과목 점수 (pandas plot)')
        plt.title('과목별 평균 점수 막대 그래프')
        plt.xlabel('과목')
        plt.ylabel('점수')
        plt.show()

    # 평균 점수 상위 5명 막대 그래프
    def score_top5_mean(self):
        # 3과목 평균 axis=1, 
        #내림차순 정렬 sort_values(ascending=False)
        # 5명만 .head(5)
        df_sorted = self.df[self.score_columns].mean(axis=1).sort_values(ascending=False).head(5)
    
        # 인덱스로 학생 이름 추출
        student_names = [self.df.iloc[idx]['students'] for idx in df_sorted.index]

        # 그래프
        plt.bar(student_names, df_sorted.values,color='dodgerblue') 
        plt.title('Method 1: 학생별 과목 점수 (pandas plot)')
        plt.title('평균 성적 상위 5명 막대 그래프')
        plt.xlabel('이름')
        plt.ylabel('점수')
        plt.show()

a= StudentScoreAnalysis()
a.score_mean()
a.score_top5_mean()
```
