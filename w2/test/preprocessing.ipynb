
# 전처리 1

```python
import pandas as pd
# 데이터 불러오기
df = pd.read_csv('train.csv')

# 결측치 확인 
print(df.info()) # age 714개 cabin 204개 Embarked 889로 결측치 존재
print('='*50) # 구분

# Age열 존재하는 결측치를 평균값으로 대체 
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Embarked 열의 결측치는 최빈값(Mode)으로 대체
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Cabin 열은 결측치가 너무 많아서 삭제 

df_del = df.drop('Cabin', axis=1) #Cabin 열 아예 제거 (원본 보존)
print('='*50) # 구분
print(df_del.info()) # 제거 확인
```

# 전처리 2

```python
import pandas as pd

df = pd.read_csv('sales.csv')
print(df.info()) #결측치 확인
print('='*50) # 구분

# 중복 데이터 개수 확인
print(f'중복데이터는 : {df.duplicated().sum()}개 입니다.')  
# 중복된 행 모두 제거. 첫 번째만 남김 (keep='first')
df_ndupl = df.drop_duplicates(keep='first') 

# len으로 존재하는 행 개수를 보여줌
print(f'중복된 행을 모두 제거한 후, 남은 데이터의 개수는 {len(df_ndupl)}개입니다.')
print('='*50) # 구분
print(df_ndupl)
```

# 전처리 3

```python
import pandas as pd

df = pd.read_csv('diamonds.csv') #파일 불러오기

print(df.info()) # 결측치 확인 (없음)
print(df.duplicated().sum()) #중복데이터 (행)

#  이상치 처리 -> 1사분위수, 3사분위수
Q1 = df['carat'].quantile(0.25) # 1사분위수
Q3 = df['carat'].quantile(0.75) # 3사분위수
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 = 데이터에 carat행이 lower_bound보다 작거나
# upper_bound 보다 크면 이상치라고 취급. 
outliers = df[(df['carat'] < lower_bound) | (df['carat'] > upper_bound)]

#print(outliers)

df_n_outliers = df[(df['carat'] >= lower_bound) & (df['carat'] <= upper_bound)] # 원본데이터 유지

# 원래 53940개 제거 후 52051
print('='*50) # 구분
print(f'초기 데이터의 개수: {len(df)}')
print(f'하한: {lower_bound}, 상한: {upper_bound}')
print(f'이상치 데이터의 개수: {len(outliers)}')
print(f'이상치 제거 후 데이터의 개수: {len(df_n_outliers)}')
print('='*50) # 구분

```

# 전처리 4

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder #레이블인코딩을 위한

df = pd.read_csv('adult.csv')
# Label Encoding -> 범주형 데이터를 정수로 매핑.
label_encoder = LabelEncoder()


#데이터 레이블 인코딩 _label 빼면 그냥 원래 데이터에 저장됨. 
for i in ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']:
    df[i+'_label'] = label_encoder.fit_transform(df[i])

print(df.info()) # 정수로 변경돼었는지 확인.

```

# 전처리 5

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# df에 csv 파일 읽어오기.
df = pd.read_csv('winequality-red.csv')

scaler_standard = StandardScaler() #객체 생성.
#'fixed acidity', 'volatile acidity', 'citric acid', 
# 'residual sugar', 'chlorides' 이 있으면 df에 테이블을 추가하는데 
# 그 안에 들어갈 데이터를 StandardScaler로 표준화
for i in ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides']:
    df[i+'_std'] =scaler_standard.fit_transform(df[[i]])
    
print(df.info())
df
```
