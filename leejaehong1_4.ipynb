```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False



class SalesAnalysis:
    def __init__(self):
        # 2024년 1월 1일부터 12월 31일까지 날짜 생성
        dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
        # 1000~10000 사이의 난수로 일별 매출 생성
        sales = np.random.randint(1000, 10001, size=len(dates))
        # DataFrame으로 저장 (self.df)
        self.df = pd.DataFrame({
            'date': dates,
            'sales': sales
        })

    def monthly_total(self):    
        return self.df.groupby(self.df['date'].dt.month)['sales'].sum()
    
    def get_monthly_lists(self):
        monthly_data = self.df.groupby(self.df['date'].dt.month)['sales'].sum()
        month_list = monthly_data.index.tolist()  # 월
        monthly_sales_list = monthly_data.values.tolist()  # 월별 매출
        
        plt.plot(month_list,monthly_sales_list,
         marker='o',
         linestyle='-', # " - " => 꺾은선 그래프
         color='green',
         label = '월 매출'
         )
        # 그래프 설정 해줘야됨.
        plt.title('2024년 월 매출표')
        plt.xlabel('월')
        plt.ylabel('매출')
        #plt.legend() #범례
        # 그래프 출력
        plt.grid(True) #격자선
        plt.show()

a = SalesAnalysis()
print(a.monthly_total())
a.get_monthly_lists()



```
